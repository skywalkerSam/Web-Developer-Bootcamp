# Normal Project-One
    >_ I'm gonna create a beautiful plot...

```python
%matplotlib inline
import matplotlib.pyplot as plt

#1 Prepare Data
x = [1, 3, 5, 7, 9, 11, 13]
y = [10, 90, 30, 180, 5, 50, 90]

#2 Setup Plot
fig, ax = plt.subplots(figsize=(9, 9)) #(Width, Height)

#3 Plot Data
ax.plot(x, y)

#4 Customize Plot
ax.set(title="Normal Plot",
      xlabel="X-axis",
      ylabel="Y-axis")

#5 Save and show figure
fig.savefig("Images/Normal_Plot.png")
```
